```python
import pandas as pd
import numpy as np
```

```python
combined_df=pd.read_csv("C:/Users/91848/Desktop/Car_Dekho/dataset/combined_df.csv")
combined_df
```

```python
threshold = 0.5
combined_df = combined_df.loc[:, combined_df.isnull().mean() < threshold]
combined_df
```

```python
combined_df.columns
```

```python
# List of columns to keep
important_columns = [
    'ft', 'bt','oem', 'model', 'variantName', 'modelYear', 'Registration Year', 'price',
    'Kms Driven', 'Fuel Type', 'Transmission', 'Engine Displacement', 'Displacement',
    'Max Power', 'Torque', 'Mileage','Engine' ,'ownerNo', 'Ownership', 'Insurance Validity', 
    'Seats', 'Length', 'Width', 'Height', 'Wheel Base', 'Kerb Weight','City'
]

# Filter combined_df to keep only the important columns
combined_df = combined_df[important_columns]
combined_df
```

```python
combined_df.isna().sum()
```

```python
# Filter rows where 'bt' column has null values
bt_null_rows = combined_df[combined_df['bt'].isnull()]
bt_null_rows
```

```python
# Replace NaN values in 'bt' with 'Minivans' for specific models
combined_df.loc[combined_df['model'] == 'Toyota Vellfire', 'bt'] = 'Minivans'
combined_df.loc[combined_df['model'] == 'Maruti Eeco', 'bt'] = 'Minivans'
```

```python
# drop nan values
combined_df.dropna(subset=["Insurance Validity","Kms Driven","Engine",],inplace=True)
combined_df
```

```python
combined_df = combined_df.drop(columns=['Ownership','Engine Displacement','Displacement','Wheel Base','Kerb Weight','Registration Year'])
```

```python
combined_df.isna().sum()
```

```python
# retrive seats nan values
combined_df[pd.isna(combined_df["Seats"])]
```

```python
# replace nan values of seats
combined_df.loc[combined_df['model']=='Maruti Estilo','Seats'] = 5.0
combined_df.loc[combined_df['model']=='Nissan Kicks','Seats'] = 5.0
combined_df.loc[combined_df['model']=='Ford Endeavour','Seats'] = 7.0
combined_df.loc[combined_df['model']=='Ambassador','Seats'] = 5.0
```

```python
combined_df.isna().sum()
```

```python
combined_df['Kms Driven'] = combined_df['Kms Driven'].astype(str).str.replace('Kms', '').str.replace(',', '').str.strip().astype(int)
combined_df['Mileage'] = combined_df['Mileage'].astype(str).str.replace('kmpl', '').str.replace('km/kg', '').str.strip().astype(float)
combined_df['Engine'] = combined_df['Engine'].str.replace('CC', '').str.strip()
combined_df['Seats'] = combined_df['Seats'].str.extract(r'(\d+)')
combined_df['Seats'] = combined_df['Seats'].fillna(0).astype(int)
combined_df['price'] = combined_df['price'].str.extract(r'(\d+\.?\d*)').astype(float)
```

```python
combined_df['Mileage'] = combined_df['Mileage'].fillna(combined_df['Mileage'].mean())
```

```python
for column in ['Length', 'Width', 'Height']:
    combined_df[column] = combined_df[column].astype(str).str.replace(' mm', '', regex=False)  # Remove ' mm' suffix
    combined_df[column] = combined_df[column].str.replace('mm', '', regex=False)  # Remove 'mm' suffix
    combined_df[column] = combined_df[column].str.strip()  # Remove leading/trailing spaces
    
    # Convert to float, handling potential NaN values
    combined_df[column] = pd.to_numeric(combined_df[column], errors='coerce')  # Convert to float, coerce errors to NaN
```

```python
# Fill missing values in Length, Width, and Height based on the median for each model
combined_df['Length'] = combined_df.groupby('model')['Length'].transform(lambda x: x.fillna(x.median()))
combined_df['Width'] = combined_df.groupby('model')['Width'].transform(lambda x: x.fillna(x.median()))
combined_df['Height'] = combined_df.groupby('model')['Height'].transform(lambda x: x.fillna(x.median()))
```

```python
combined_df.isna().sum()
```

```python
combined_df['Max Power'] = pd.to_numeric(combined_df['Max Power'].str.extract(r'(\d+)')[0], errors='coerce')
combined_df['Torque'] = pd.to_numeric(combined_df['Torque'].str.extract(r'(\d+)')[0], errors='coerce')
```

```python
print(combined_df[['Max Power','Torque']])
```

```python
# Fill null Max Power based on Engine Displacement
combined_df['Max Power'] = combined_df.groupby('Engine')['Max Power'].transform(
    lambda x: x.fillna(x.mean())
)

# Fill null Torque based on Engine Displacement
combined_df['Torque'] = combined_df.groupby('Engine')['Torque'].transform(
    lambda x: x.fillna(x.mean())
)
```

```python
combined_df.isna().sum()
```

```python
combined_df = combined_df.dropna(subset=['Max Power', 'Torque', 'Width'])
```
